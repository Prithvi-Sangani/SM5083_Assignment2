```python
import math
import matplotlib.pyplot as plt
xab = [2,3]
yab = [3,4]
n1 = ["$A$","$B$"]

xcd = [3,5,2.7142857142857144,2]
ycd = [3,-2,3.7142857142857144,11/2]
n2 = ["$C$","$D$","$X$"]
fig, ax = plt.subplots()
ax.scatter(xab, yab)
plt.plot(xab,yab)
for i, txt in enumerate(n1):
    ax.annotate(txt, (xab[i], yab[i]))
ax.scatter(xcd, ycd)
plt.plot(xcd,ycd)
for i, txt in enumerate(n2):
    ax.annotate(txt, (xcd[i], ycd[i]))
plt.xlim(1.5, 5.5), plt.ylim(-2.5, 6)
plt.grid()
plt.show()
```
